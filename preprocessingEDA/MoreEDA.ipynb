```python
df = pd.read_csv("RussianHousingTrain.csv")
df.isnull().sum()
```

```python
df_categorical.columns.values
df_num.columns.values
```

```python
import missingno as msno
```

```python
cols = df.columns[:30]
msno.matrix(df[cols])
```

```python
plt.figure(figsize=(16, 12))
sns.heatmap(df[cols].isnull())
```

```python
for col in df.columns:
    missing = np.mean(df[col].isnull())
    # print(str(col) + " - " + str(round(missing*100)) + '%')
    print('{} - {}%'.format(col, round(missing*100)))
```

```python
df[df_num.columns] = df[df_num.columns].fillna(df[df_num.columns].median())
```

```python
low_info_cols = []
for col in df.columns:
    percent = (df[col].value_counts() / len(df.index)).iloc[0]

    if percent > 0.95:
        low_info_cols.append(col); print('{0}: {1:.5f}%'.format(col, percent*100))


```
